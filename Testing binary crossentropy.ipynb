```python
import keras
from keras.models import Model, Sequential
import numpy as np
from keras.optimizers import SGD, Adam
from keras.layers import Dense, LSTM, Input, Reshape, merge
```

```python
def get_discriminative_model():
    model = Sequential()
    model.add(Dense(3, input_dim=1, init='uniform', activation='relu'))
    model.add(Dense(1, init='uniform', activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
    return model
```

```python
discriminator = get_discriminative_model()
```

```python
inputs = np.array([20, 3, 40, 5, 60, 7, 80, 9])
outputs = np.array([1, 0 , 1, 0, 1, 0, 1, 0])
inputs2 = np.array([4,5,6])
```

```python
weights = discriminator.get_weights()
print weights

history = discriminator.fit (inputs, outputs, nb_epoch=100)
print discriminator.get_weights()
predictions = discriminator.predict(inputs2)
```
